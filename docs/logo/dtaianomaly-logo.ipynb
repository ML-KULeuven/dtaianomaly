```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
```

```python
nb_samples = 1000
x = np.linspace(0, 1, 1000)

bell_curve = scipy.stats.norm.pdf((x-0.5) * 8, 0, 1) 
wave = np.sin(2 * np.pi * x * 5)

fig, axs = plt.subplots(1, 3, figsize=(10, 3))
axs[0].plot(bell_curve)
axs[1].plot(wave)
axs[2].plot(bell_curve * wave);
```

```python
plt.figure(figsize=(6, 5))
plt.plot(bell_curve * wave, c='#00407A', linewidth=20)
plt.axis('off')
plt.tight_layout()
plt.savefig('ts.svg', transparent=True)
```
